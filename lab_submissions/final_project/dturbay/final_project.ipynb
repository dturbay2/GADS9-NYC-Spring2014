For this project, I have attempted to predict the winner of the World Cup. With the increasing precense of data in sports, this has become a popular exercise. However, soccer is lagging in the data revolution for reasons that are intrinsic to the sport and a lack of belief in the general soccer community. It is much more difficult to quantify a player's value in soccer in comparison to other sports; goals are too scarce, and players can add value in ways that cannot be captured by the usual metrics of goals, assists, blocks, passes. That being said, it is difficult to find good, usable data sets to work off.

Most predicors that I have seen, rely on historical data. The models that I've come across use data from world cups that date as far back as 1950! It's pretty obvious that a country's current performance is completely independent from their performance in the 1950s. Because of this, these models are very skewed towards Brazil because of their incomparable historical results.

To build a more accurate prediction, we need to consider recently played games (at least within the last 10 years. I think a team's history can help but probably not that much). Although I generally dont think FIFA rankings have much meaning, partly because there is no transative property in soccer (A beats B and B beats C does not imply that A beats C), it is a good aggregation of recent performance and it provides a quantitative way for separating teams at a given point in time.

The main problem for predicting how team A will do against team B is that the scale of international/intercontinental soccer games is extremely limited. Aside from the WC every four years, you have a few friendlies here and there over the course of a year. Even so, friendlies are a very bad indicator of a team's strength. To overcome this, I was planning on taking into account data from the WC qualifiers (WCQ) which have many more games that actually matter, and finding a way of linking teams between continents. In order to link different teams, I was planning on grouping/segmenting countries:

   * Label every country by tier or strength. this can relate to segmentation based on ranking. instead of thinking about, how will X do against Y, think about how will X do against a team in the category of Y
    
   * categories can be based on prestige, history, continent, player quality, coach, ranking distribution (top 10%, etc...); can be qualitative or quantitative
        
Unfortunately, I had more ambitious plans than I had time, so I was not able to follow my plan fully. Getting WCQ results in a usable format was much harder than expected, so I was forced to abandon that idea for the sake of time, and just use more data from previous world cups). I ended up getting WC results from 2002, 2006, and 2010, along with Rankings for those years in the month before the WC.

In the end I built a data set with all group stage games in the past three WCs, and joined the repsective ranks for every time at the given time. I used the FIFA rankings for each team in a game to predict the goal difference of the game (Linear Regression) and to predict the outcome regarless of the score (Logistic Regression). Neither model proved to be too great of a fit, but the logfit was slighlty better.

Below are helper functions that I wrote to read txt files and generate dataframe. Another helper function joins WC result data to rankings data

```python
#now delete everything from the list that has length 1

def remove_spaces(df,column):
    for entry in df[column]:
        for elem in entry:
            if len(elem) < 1:
                entry.remove(elem)
                remove_spaces(df,column) #made recursive so it removes all spaces and doest short circuit
                
#this generates rank df for a given year
def rank_gen(ranktxt,year):
    ranktxt.rename(columns={'# Team Points Change':'ranks'},inplace=True)
    ranktxt['rank'] = ranktxt['ranks'].map(lambda x: x.split(' ')[0])
    remove_spaces(ranktxt,'rank')
    ranktxt['team1'] = ranktxt['ranks'].map(lambda x: x.split(' ')[1])
    ranktxt.drop('ranks',axis=1,inplace=True)
    ranktxt['team2'] = ranktxt['team1']
    ranktxt['year'] = year
    return ranktxt


def rank_gen2(ranktxt, year):
    ranktxt = ranktxt[['#','Team']]
    ranktxt.rename(columns={'#':'rank','Team':'team1'},inplace=True)
    ranktxt['team2'] = ranktxt['team1']
    ranktxt['year'] = year
    return ranktxt

#function that takes in realtively cleaned txt, and parses into df with
#columns with each time their score and score diff for each time

def df_gen(txtname,name):
    txtname.rename(columns={txtname.columns[0]:'test_col'},inplace=True)
    
    #for every row, we can split up each section and we only need the 2nd (index 1 element of the list)
    txtname['game'] = txtname['test_col'].map(lambda x: x.split('   ')[1:-1])
    
    remove_spaces(txtname,'game') # use function above to clean up
    
    txtname = txtname[txtname.test_col.map(len) > 8] # this removes the group name rows and just leaves games!
    
    txtname['team1gen'] = txtname['game'].map(lambda x: ' '.join(str(y) for y in x).split('-')[0]) # 'south africa 1'
    txtname['team2gen'] = txtname['game'].map(lambda x: ' '.join(str(y) for y in x).split('-')[1])# '1 mexico'
    
    txtname['team1'] = txtname['team1gen'].map(lambda x: x[:-1].strip())
    txtname['team2'] = txtname['team2gen'].map(lambda x: x[1:].strip())
    
    txtname['score_team1'] = txtname['team1gen'].map(lambda x: int(x[-1]))
    txtname['score_team2'] = txtname['team2gen'].map(lambda x: int(x[0]))
    
    txtname['score_diff_1'] = txtname.score_team1 - txtname.score_team2
    txtname['score_diff_2'] = txtname.score_team2 - txtname.score_team1
    
    #clean columns
    txtname.drop(['test_col','game','team1gen','team2gen'],axis=1,inplace=True)
    
    #Now let's add the competition name and year:
    txtname['competition'] = name
    
    txtname.reset_index(drop=True,inplace=True)
    return txtname


def merge_rank(results_df,rank_df):
    #merge rank of team1
    results_df = pd.merge(results_df, rank_df,on='team1')
    results_df.drop(['team2_y','year'],axis=1,inplace=True)
    results_df.rename(columns={'team2_x':'team2','rank': 'team1_rank'},inplace=True)
    #merge rank of team2
    results_df = pd.merge(results_df, rank_df,on='team2')
    results_df.drop(['team1_y','year'],axis=1,inplace=True)
    results_df.rename(columns={'team1_x':'team1','rank': 'team2_rank'},inplace=True)
    return results_df
```

```python
def get_winner(t1,t2):#pass principal team as a string for t1
    # we want to look up the rankings of team one (t1) and team 2 (t2) in the most recent ranking data
    t1_rank = rankings[rankings['country'] == t1]['fifa_rank'].values[0]
    t2_rank = rankings[rankings['country'] == t2]['fifa_rank'].values[0]
    
    # once we have rankings we can predict with our logclf whether t1 will lose, win, or, draw
    outcome = logclf.predict([t1_rank,t2_rank])[0]
    
    # it can return the output of the classifier, aka the number of pts from that game. This can be used to get pts for t2
    if outcome == 3:
        winner = t1
    elif outcome == 0:
            winner = t2
    else: 
        winner = 'tie'
        
    # the output will be the winning team
    return winner #, t1_rank, t2_rank, outcome

print 'Colombia vs. Brazil: ',get_winner('Colombia','Brazil')
print 'Colombia vs. Greece: ',get_winner('Colombia','Greece')
```

```python
#this modifies the table that gets passed and adds column with total points
def mk_table(group):
    #create list of the winner of each game
    results = [] 
    for game in group_games(group):
        results = results + [get_winner(game[0],game[1])]
    
    #create column with the number of points for each team
    group['points'] = group['country'].map(lambda x: results.count(x)*3)
    group = group[['country','points']]
    return group
```
