```python
        import pandas as pd
        import numpy as np
        from matplotlib import pyplot as plt
        from sklearn.linear_model import LinearRegression
        from sklearn.feature_selection import f_regression
        from sklearn.metrics import mean_squared_error, mean_absolute_error
        %pylab inline
        
        data = pd.read_csv('./data/day.csv')
        data.head()
        print data.columns
        data.head()
```

```python
#Casual Data

#make vector with causal usage

casual = data.casual
reg = data.registered
cnt = data.cnt

feeltemp = data.atemp



fig = plt.figure()
fig.set_figwidth(20.0)
fig.set_figheight(8.0)

ax1 = fig.add_subplot(2,4,1) # one row, two columns, first plot
ax1.hist(data.atemp)
plt.title('atemp')

ax2 = fig.add_subplot(2,4,2) # one row, two columns, second plot
ax2.hist(data.mnth)
plt.title('monthly')

ax3 = fig.add_subplot(2,4,3)
ax3.hist(data.windspeed)
plt.title('windspeed')

ax4 = fig.add_subplot(2,4,4)
ax4.hist(data.hum)
plt.title('humidity')

ax5 = fig.add_subplot(2,4,5)
ax5.scatter(np.log(data.atemp),np.log(casual))
plt.title('atemp')

ax6 = fig.add_subplot(2,4,6)
ax6.scatter(data.instant,casual)
plt.title('overtime')

ax7 = fig.add_subplot(2,4,7)
ax7.scatter(data.windspeed,casual)
plt.title('windspeed')

ax8 = fig.add_subplot(2,4,8)
ax8.scatter(data.hum,casual)
plt.title('humidity')

print fig
```

```python
#whats better atemp or temp? find p vals

lin_fit = LinearRegression()
log_fit = LinearRegression()

lin_fit.fit(data[['atemp','hum','windspeed']].values,data['casual'].values)
log_fit.fit(np.log(data[['atemp']].values),np.log(data['casual'].values))

print 'lin inter:', lin_fit.intercept_
print 'lin coeff:', lin_fit.coef_

print 'log inter:', log_fit.intercept_
print 'log coeff:', log_fit.coef_
```

```python
lin_fit.score(data[['atemp','hum','windspeed']].values,data['casual'].values)
```

```python
f_score, p_score = f_regression(data[['atemp','hum','windspeed']].values,data['casual'].values)
print "p-scores (atemp, humidity, windspeed):", p_score
```

```python
##########
#questions
#seems like im getting extremely low pvalues and a pretty bad R^2 val... 
```

```python
fig2 = plt.figure()
fig2.set_figwidth(20)
axis1 = fig2.add_subplot(1,2,1)
plt.scatter(data.instant, casual)
plt.title('Casual Data')

axis2 = fig2.add_subplot(1,2,2)
plt.scatter(data.instant,lin_fit.predict(data[['atemp','hum','windspeed']].values))
plt.title('Predicted Data')
```

```python
plt.plot(sort(lin_fit.predict(data[['atemp','hum','windspeed']].values)))
```

```python
#run regressions for registered (reg) and cnt data
feats = data[['atemp','hum','windspeed']]

regfit = LinearRegression()
cntfit = LinearRegression()

regfit.fit(feats,reg)
cntfit.fit(feats,cnt)

print 'reg inter:', regfit.intercept_
print 'reg coeff:', regfit.coef_

print 'cnt inter:', cntfit.intercept_
print 'cnt coeff:', cntfit.coef_
```

```python
#find p values for each

freg,preg = f_regression(feats,reg)
fcnt,pcnt = f_regression(feats,cnt)

print "p val for registered: ", preg
print "p val for cnt: ", pcnt
```

```python
#find R^2
Rreg = regfit.score(feats,reg)
Rcnt = cntfit.score(feats,cnt)

print 'R squared for registered: ', Rreg

print 'R squared for cnt: ', Rcnt
```

```python
#we see that model is much better for cnt since it Rsqr and p vals are lower
```

```python
fig3 = plt.figure()
fig3.set_figwidth(20.0)
fig3.set_figheight(8.0)
xax = data.instant


ax1 = fig3.add_subplot(3,3,1)
plt.title('Registered Actual')
plt.scatter(xax,reg)

ax2 = fig3.add_subplot(3,3,2)
plt.title('Registered Predicted')
plt.scatter(xax,regfit.predict(feats))

ax3 = fig3.add_subplot(3,3,3)
plt.title('Registered linear')
plt.plot(sort(reg),sort(regfit.predict(feats)))


ax4 = fig3.add_subplot(3,3,4)
plt.title('Casual Actual')
plt.scatter(xax,casual)


ax5 = fig3.add_subplot(3,3,5)
plt.title('Casual Predict')
plt.scatter(xax,lin_fit.predict(feats))

ax6 = fig3.add_subplot(3,3,6)
plt.title('Casual Linear')
plt.plot(sort(casual),sort(lin_fit.predict(feats)))


ax7 = fig3.add_subplot(3,3,7)
plt.title('CNT Actual')
plt.scatter(xax,cnt)

ax8 = fig3.add_subplot(3,3,8)
plt.title('CNT Predicted')
plt.scatter(xax,cntfit.predict(feats))

ax9 = fig3.add_subplot(3,3,9)
plt.title('CNT Linear')
plt.plot(sort(cnt),sort(cntfit.predict(feats)))

print fig3
```
